# GSE131907 LUAD TME Reanalysis Overview

Interactive companion notebook for inspecting the generated metadata, composition, signature, subtype, and paired-patient outputs. The reproducible workflow lives in `scripts/`; this notebook is intentionally lightweight.

```python
from pathlib import Path
import pandas as pd

project = Path('..').resolve()
docs_tables = project / 'docs' / 'tables'
meta = pd.read_csv(project / 'data/metadata/gse131907_cell_metadata_clean.csv')
meta.shape
```

## Tissue-Site and Cell-Type Composition

```python
pd.crosstab(meta['tissue_site'], meta['cell_type_clean'])
```

```python
cell_counts = pd.read_csv(docs_tables / 'cell_counts_by_sample.csv')
cell_counts.sort_values('n_cells', ascending=False).head(12)
```

## Signature Contrasts

```python
contrasts = pd.read_csv(docs_tables / 'signature_site_contrasts.csv')
cols = ['cell_type_clean', 'signature', 'contrast', 'n_contrast_samples', 'n_reference_samples', 'delta_mean', 'padj_bh']
contrasts[cols].sort_values(['padj_bh', 'delta_mean'], ascending=[True, False]).head(12)
```

## Cell-Subtype Summaries

```python
subtypes = pd.read_csv(docs_tables / 'subtype_composition_by_tissue.csv')
subtypes.sort_values(['tissue_site', 'fraction'], ascending=[True, False]).groupby('tissue_site').head(5)
```

```python
subtype_scores = pd.read_csv(docs_tables / 'signature_scores_by_tissue_subtype.csv')
subtype_scores.sort_values('HYPOXIA', ascending=False).head(12)
```
